```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
```

```python
final  = pd.read_csv('storesalesdata.csv')
final.head()
```

```python
nf = final.groupby(['Store', 'Year'], as_index=False).sum()
nf.head()
```

```python
final = nf[nf['Year'] == 2015]
final.head()
```

```python
final['Total Sale'] = final['Dry_Grocery'] + final['Dairy'] + final['Frozen_Food'] + final['Meat'] + final['Produce']+ final['Floral']+ final['Deli']+ final['Bakery']+ final['General_Merchandise']
```

```python
final['Ptc_Dry_Grocery'] = final['Dry_Grocery'] / final['Total Sale'] * 100
final['Ptc_Dairy'] = final['Dairy'] / final['Total Sale'] * 100
final['Ptc_Frozen_Food'] = final['Frozen_Food'] / final['Total Sale'] * 100
final['Ptc_Meat'] = final['Meat'] / final['Total Sale'] * 100
final['Ptc_Produce'] = final['Produce'] / final['Total Sale'] * 100
final['Ptc_Floral'] = final['Floral'] / final['Total Sale'] * 100
final['Ptc_Deli'] = final['Deli'] / final['Total Sale'] * 100
final['Ptc_Bakery'] = final['Bakery'] / final['Total Sale']* 100
final['Ptc_General_Merchandise'] = final['General_Merchandise'] / final['Total Sale'] * 100
```

```python
final.drop(['Day',	'Month',	'Dry_Grocery'	,'Dairy'	,'Frozen_Food',	'Meat','Produce','Floral',	'Deli',	'Bakery',	'General_Merchandise'], axis=1,inplace=True)
```

```python
new = df1.merge(final, on='Store')
```

```python
new.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler,Normalizer,minmax_scale,PowerTransformer,QuantileTransformer
```

```python
fnew = new.iloc[ :, -9:]
#scaler = QuantileTransformer()
scaler = stats.zscore(fnew,ddof=1)
```

```python
fnew[['Ptc_Dry_Grocery','Ptc_Dairy','Ptc_Frozen_Food','Ptc_Meat','Ptc_Produce','Ptc_Floral','Ptc_Deli','Ptc_Bakery','Ptc_General_Merchandise']] = scaler.fit_transform(fnew[['Ptc_Dry_Grocery','Ptc_Dairy','Ptc_Frozen_Food','Ptc_Meat','Ptc_Produce','Ptc_Floral','Ptc_Deli','Ptc_Bakery','Ptc_General_Merchandise']])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,
                random_state=3)
kmeans.fit(scaler)
y_kmeans = kmeans.predict(scaler)
```

```python
new['cluster'] = pd.DataFrame(y_kmeans)
```
